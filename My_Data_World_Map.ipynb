## Section 1 — What is Data Analytics?

Data analytics is the systematic process of examining raw data to uncover patterns, draw conclusions, and support informed decision-making in organisations.

It sits at the intersection of statistics, computer science, and business thinking. A data analyst asks questions of data that an organisation cannot easily answer by looking at it raw — and then communicates those answers to people who need to act on them.

### The Four Types of Analytics

| Type | Question it answers | Example |
|---|---|---|
| **Descriptive** | *What happened?* | Monthly revenue report by region |
| **Diagnostic** | *Why did it happen?* | Sales fell because supply was delayed by 3 weeks |
| **Predictive** | *What will happen?* | 23% of customers are likely to cancel next month |
| **Prescriptive** | *What should we do?* | Offer a 15% discount to the at-risk 23% to retain them |

The work in this notebook is primarily **descriptive and diagnostic** — I am mapping what data I generate and beginning to understand why organisations collect it.

## Section 4 — Non-Obvious Insights From My Data World

The obvious insight is "apps collect my data." The interesting insights are about what that data actually enables — and what it reveals about the power dynamic between the user and the platform.

### Insight 1: I am the product, even when I pay
Netflix costs £10.99/month. Spotify costs £9.99/month. In both cases I am *paying* for the service, yet both companies still collect extensive behavioural data about me. I assumed that paying for something meant I was not being monetised — I was wrong. My behaviour is used to improve their algorithm and inform commissioning decisions worth hundreds of millions of pounds.

### Insight 2: The data is more valuable in aggregate than individually
No single data point about me is interesting. But when my Google Maps data is combined with 5 million other users, it becomes a real-time traffic prediction system that Google sells access to through its advertising platform. My individual data is raw material. At scale, it becomes a product.

### Insight 3: One company controls more of my data than I realised
Google Maps + YouTube + Google Search = three of my top twelve daily apps, all owned by Alphabet. Alphabet can see where I go (Maps), what I watch (YouTube), and what I think about (Search) — and cross-reference all three. The profile that creates is far more powerful than any single source.

### Insight 4: Not all data collection is equal — Apple's model is different
Apple Health explicitly does not sell my health data. It stores it on-device or in end-to-end encrypted iCloud. This is a deliberate commercial choice — Apple uses privacy as a competitive differentiator. This shows that data collection is not inevitable; it is a *business model decision*.

### Insight 5: My "free" time costs me my privacy
WhatsApp is free. Instagram is free. YouTube (basic) is free. Combined, these three apps collect my social graph, my content preferences, my message metadata, and my viewing behaviour. The business model is clear: if you are not paying for the product, the data you generate *is* the product being sold.

````python
# ── ANALYTICAL EXTENSION: Map the data flow for one app (Netflix) ────────────
# This demonstrates the full journey data takes — from my behaviour to
# a business decision — using Netflix as a worked example.

netflix_data_flow = {
    "Step 1 — Collection": {
        "what": "I watch 'The Crown' for 45 minutes, then pause.",
        "data_captured": [
            "Title: The Crown, Season 3, Episode 4",
            "Watch duration: 45 min 12 sec of 58 min total",
            "Paused at: 45:12 (78% completion)",
            "Device: iPhone 15, 4G connection, 22:14 on Tuesday"
        ]
    },
    "Step 2 — Storage": {
        "what": "Data is written to Netflix's distributed database (likely AWS S3/DynamoDB)",
        "data_captured": ["Stored with my account ID, device fingerprint, session token"]
    },
    "Step 3 — Processing": {
        "what": "Batch processing job analyses my behaviour against 230M other users",
        "data_captured": [
            "Users who watched 78% of S3E4 are 67% likely to continue the series",
            "Users who pause at night are 43% more likely to return the next day"
        ]
    },
    "Step 4 — Decision": {
        "what": "Algorithm updates my recommendation weights",
        "data_captured": [
            "Historical drama weighting: increased +12%",
            "Next session: The Crown S3E5 shown as first recommendation"
        ]
    },
    "Step 5 — Business Impact": {
        "what": "Netflix reports to its board that churn is reduced by personalisation",
        "data_captured": [
            "Estimated: personalisation saves Netflix £1.5bn/year in reduced churn",
            "Source: Netflix internal estimate cited in their engineering blog"
        ]
    }
}

print("DATA FLOW ANALYSIS — Netflix (one viewing session)")
print("=" * 60)
for step, details in netflix_data_flow.items():
    print(f"\n{step}")
    print(f"  What happens: {details['what']}")
    print(f"  Data involved:")
    for item in details["data_captured"]:
        print(f"    → {item}")
    
print()
print("Key takeaway: One 45-minute viewing session generates ~15 data points")
print("that feed directly into a system managing £1.5bn in business value.")
print("This is what data analytics looks like at scale.")
````
